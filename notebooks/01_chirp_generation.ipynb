```python
# Cell 1
"""
Chirp Generation Demo
=====================
Demonstrates LFM chirp signal generation and visualization.
"""

import sys
sys.path.append('../src')

import numpy as np
import matplotlib.pyplot as plt
from chirp import generate_chirp
from config import CONFIG

# Cell 2
# Generate chirp
t, chirp = generate_chirp(CONFIG)

# Cell 3
# Plot time domain
plt.figure(figsize=(10, 3))
plt.plot(t*1e6, np.real(chirp), label='Real')
plt.plot(t*1e6, np.imag(chirp), label='Imag', alpha=0.7)
plt.xlabel('Time (Î¼s)')
plt.ylabel('Amplitude')
plt.title('LFM Chirp Signal')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Cell 4
# Plot frequency domain
freq = np.fft.fftfreq(len(chirp), t[1]-t[0])
chirp_fft = np.fft.fft(chirp)

plt.figure(figsize=(10, 3))
plt.plot(freq/1e6, np.abs(chirp_fft))
plt.xlabel('Frequency (MHz)')
plt.ylabel('Magnitude')
plt.title('Chirp Spectrum')
plt.grid(alpha=0.3)
plt.xlim(-100, 100)
plt.show()
```
